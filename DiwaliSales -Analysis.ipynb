# State 

```python
data.head(1)
```

```python
data.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State',y='Orders',data=data)
plt.show()
```

```python
#Top 10 states find the total number of orders

sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Orders',data=sales_state)
plt.show()
```

```python
#find the total no orders per state

dx=sns.set(rc={'figure.figsize':(20,10)})
dx=sns.barplot(x='State', y='Orders',data=sales_state )


for bars in dx.containers:
    dx.bar_label(bars)
```

```python
#Top 10 states find the total number of Amount 
sales_state=data.groupby(['State'],as_index=False)['TotalAmount'].sum().sort_values(by='TotalAmount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='TotalAmount',data=sales_state)
plt.show()
```

```python
#find the total amount per state

dx=sns.set(rc={'figure.figsize':(15,5)})
dx=sns.barplot(x='State',y='TotalAmount',data=sales_state )


for bars in dx.containers:
    dx.bar_label(bars)
```

From above graphs we can see that most of the orders and total amount are from UP ,Maharashta ,Karnataka
